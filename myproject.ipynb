```python
import requests

all_data = []


for page in range(1, 6):

    url =f"https://api.coingecko.com/api/v3/coins/markets?vs_currency=inr&per_page=250&order=market_cap_desc&page={page}&sparkline=False"

    print(f"Fetching Page {page}...")

    response = requests.get(url)

    if response.status_code == 200:
        data = response.json()
        all_data.extend(data)   # Add page data
    else:
        print("Error:", response.status_code)
        break

print("Total Data Collected:", len(all_data))
```

```python
import pandas as pd
df = pd.DataFrame(all_data)
df.to_csv("cryptodata.csv", index=False)
```

**Top 3 Coins – Historical Prices **


```python
import requests
from datetime import datetime
import pandas as pd
hist_prices=[]
coins_id = ["bitcoin","ethereum", "tether"]
for coin_id in coins_id:
 url = f"https://api.coingecko.com/api/v3/coins/{coin_id}/market_chart?vs_currency=usd&days=365"
 print(f"Fetching data for {coin_id} ...")
 response = requests.get(url)

 data=response.json()
 print(data)
 df = pd.DataFrame(data["prices"], columns=[ "date", "price_usd"])

 #df["date"] = pd.to_datetime(df["date"]).dt.date
 #df["date"] = pd.to_datetime(df["date"])
 df["date"] = pd.to_datetime(df["date"], unit="ms").dt.date

 df["coin_id"] = coin_id
 hist_prices.append(df)

hist_df = pd.concat(hist_prices)
print(hist_df)

```

```python
import pandas as pd

# Load data
oil_df = pd.read_csv("https://raw.githubusercontent.com/datasets/oil-prices/main/data/wti-daily.csv")

# Convert Date column to datetime
oil_df["Date"] = pd.to_datetime(oil_df["Date"])

# Filter Jan 2020 to Jan 2026
filtered_oil_df = oil_df[
    (oil_df["Date"] >= "2020-01-01") &
    (oil_df["Date"] <= "2026-01-31")
]

```

**Stock Prices (Yahoo Finance API)**


```python
import yfinance as yf
import pandas as pd

# Choose indices
tickers = ["^GSPC", "^IXIC", "^NSEI"]
start_date = "2020-01-01"
end_date = "2026-02-09"

# Download historical data
stocks_df = yf.download(tickers, start=start_date, end=end_date, group_by="ticker")
stocks_df.head()
```

```python
stocks_df['^GSPC'].reset_index()
```

```python
stocks_df1=stocks_df['^GSPC'].reset_index().dropna()
stocks_df1["Ticker"] = "^GSPC"
```

```python
stocks_df1
```

```python
stocks_df2=stocks_df['^IXIC'].reset_index().dropna()
stocks_df2["Ticker"] = "^IXIC"
```

```python
stocks_df3=stocks_df['^NSEI'].reset_index().dropna()
stocks_df3["Ticker"] = "^NSEI"
```

```python
final_df = pd.concat([stocks_df1,stocks_df2,stocks_df3], ignore_index=True)
final_df
```

```python
final_df.to_csv("stock_prices.csv", index=False)
```
