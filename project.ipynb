```python
pd.read_sql("""
SELECT 
    Date,
    High,
    Low,
    (High - Low) AS price_difference
FROM stock_prices
WHERE Ticker = '^GSPC'
ORDER BY price_difference DESC
LIMIT 5
""", engine)
```

query 4 Get monthly average closing price for each ticker



```python
pd.read_sql("""
SELECT 
    Ticker,
    YEAR(Date) AS year,
    MONTH(Date) AS month,
    AVG(Close) AS avg_monthly_close
FROM stock_prices
GROUP BY 
    Ticker, YEAR(Date), MONTH(Date)
ORDER BY 
    Ticker, year, month
""", engine)
```

    Join Queries (Cross-Market Analysis)


Compare Bitcoin vs Oil average price in 2025.


```python
pd.read_sql("""
SELECT 
    AVG(hp.price_usd) AS btc_avg,
    AVG(op.price) AS oil_avg
FROM hist_prices hp
JOIN oil_price op 
    ON hp.date = op.date
WHERE hp.coin_id = 'bitcoin'
""", engine)
```

query 2 Check if Bitcoin moves with S&P 500 (correlation idea).


```python
df=pd.read_sql("""
SELECT 
    hp.date,
    hp.price_usd AS btc_price,
    sp.Close AS sp500_close
FROM hist_prices hp
JOIN stock_prices sp
    ON DATE(hp.date) = sp.Date
WHERE hp.coin_id = 'bitcoin'
AND sp.Ticker = '^GSPC'
ORDER BY hp.date
""", engine)

df['btc_return'] = df['btc_price'].pct_change()
df['sp500_return'] = df['sp500_close'].pct_change()

correlation = df[['btc_return', 'sp500_return']].corr()

print("Correlation:", correlation)
```

Query 4 Find days when oil price spiked and compare with Bitcoin price change.

```python
pd.read_sql("""SELECT 
    o1.date AS market_date,
    o1.price AS oil_price,
    (o1.price - o2.price) AS oil_price_spike,
    hp.price_usd AS bitcoin_price
FROM oil_price o1
JOIN oil_price o2 
    ON o1.date = DATE_ADD(o2.date, INTERVAL 1 DAY)
JOIN hist_prices hp
    ON DATE(hp.date) = o1.date
WHERE hp.coin_id = 'bitcoin'
ORDER BY oil_price_spike DESC
LIMIT 20""",engine)
```

query 5 Compare top 3 coins daily price trend vs Nifty (^NSEI).


```python
pd.read_sql("""
SELECT hp.coin_id,
       DATE(hp.date) AS date,
       hp.price_usd,
       sp.Close AS nifty_close
FROM hist_prices hp

JOIN (
    SELECT id
    FROM records
    ORDER BY market_cap DESC
    LIMIT 3
) top_coins
ON hp.coin_id = top_coins.id

JOIN stock_prices sp
ON DATE(hp.date) = sp.Date

WHERE sp.Ticker = '^NSEI'
ORDER BY date
""", engine)
```

Query 9 Join top 3 crypto coins with stock indices for 2025


```python
pd.read_sql("""
SELECT hp.coin_id,
       hp.date,
       hp.price_usd,
       sp.ticker,
       sp.close
FROM hist_prices hp

JOIN (
    SELECT id
    FROM records
    ORDER BY market_cap DESC
    LIMIT 3
) top_coins
ON hp.coin_id = top_coins.id

JOIN Stock_prices sp
ON hp.date = sp.date

WHERE YEAR(hp.date) = 2025
ORDER BY hp.coin_id, hp.date""",engine)
```

Query 10 Multi-join: stock prices, oil prices, and Bitcoin prices for daily comparison

```python
pd.read_sql("""
SELECT 
    DATE(hp.date) AS date,
    hp.price_usd AS bitcoin_price,
    op.price AS oil_price,
    sp.Close AS sp500_close
FROM hist_prices hp

JOIN oil_price op
    ON DATE(hp.date) = op.date

JOIN stock_prices sp
    ON DATE(hp.date) = sp.Date

WHERE hp.coin_id = 'bitcoin'
AND sp.Ticker = '^GSPC'

ORDER BY date
""", engine)
```
